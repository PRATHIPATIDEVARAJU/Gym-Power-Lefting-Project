## Data Visualization

```python
data.head()
```

```python
data = pd.read_csv(r"C:\Users\devad\Downloads\powerlifting_dataset.csv")
```

```python
deva = data.groupby('Weight Class')['Amount Lifted (kg)'].median()

plt.figure(figsize=(12, 6))
deva.plot(kind = 'line', marker='D', color='orange', mec='r', ms = 10)
plt.xlabel('Weight Class')
plt.ylabel('Median Amount Lifted (kg)')
plt.title('Median Weight Lifted by Weight Class', color = 'g')
plt.show()
```

```python
weight = data.groupby('Age')['Amount Lifted (kg)'].mean()

plt.figure(figsize=(12,8))
weight.plot(kind = 'line',marker = 'D',color = 'green', mec = 'r', ms = 10)
plt.xlabel('Age')
plt.ylabel('Average Amount Lifted (kg)')
plt.title('Average Weight Lifted by Age')
plt.show()
```

```python
lift_count = data['Lift Type'].value_counts()
plt.figure(figsize=(12, 8))
plt.bar(lift_count.index, lift_count.values, color = 'k')
plt.xlabel('Lift Type', color = 'b')
plt.ylabel('Count', color = 'g')
plt.title('Count of Lift Types by Lifter Name', color = 'r')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_lift_by_type = data.groupby('Lift Type')['Amount Lifted (kg)'].mean()
avg_lift_by_type.plot(kind='bar')
plt.xlabel('Lift Type')
plt.ylabel('Average Amount Lifted (kg)')
plt.title('Average Amount Lifted by Lift Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
weight_class_counts = data['Weight Class'].value_counts()
weight_class_counts.plot(kind='bar')
plt.xlabel('Weight Class')
plt.ylabel('Number of Lifters')
plt.title('Number of Lifters by Weight Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['Lifter Name'], data['Amount Lifted (kg)'])
plt.xlabel('Lifter Name')
plt.ylabel('Amount Lifted (kg)')
plt.title('Amount Lifted by Each Lifter')
plt.xticks(rotation=90)
plt.show()
```

```python
lift_count = data.groupby(['Age', 'Lift Type']).size().unstack()

plt.figure(figsize=(12, 8))
lift_count.plot(kind='bar', stacked=True,color=['skyblue', 'lightgreen', 'orange'])
plt.xlabel('Age')
plt.ylabel('Count of Lift Types')
plt.title('Count of Lift Types by Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Amount Lifted (kg)'], bins=10, edgecolor='black')
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Frequency')
plt.title('Distribution of Amount Lifted')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data, x='Amount Lifted (kg)', hue='Lift Type', multiple='stack', palette='viridis')
plt.title('Distribution of Amount Lifted by Lift Type')
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data.boxplot(column='Amount Lifted (kg)', by='Weight Class', grid=False)
plt.xlabel('Weight Class')
plt.ylabel('Amount Lifted (kg)')
plt.title('Box Plot of Amount Lifted by Weight Class')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data.boxplot(column='Amount Lifted (kg)', by='Age', grid=False)
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.title('Box Plot of Amount Lifted by Age')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Amount Lifted (kg)'])
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.title('Amount Lifted vs Age')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
lift_type_counts = data['Lift Type'].value_counts()
plt.pie(lift_type_counts, labels=lift_type_counts.index, autopct='%1.1f%%', colors=plt.get_cmap('tab20').colors)
plt.title('Distribution of Lift Types')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Amount Lifted (kg)', hue='Weight Class', palette='Set1')
plt.title('Scatter Plot of Amount Lifted vs Age')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data['Age'], data['Amount Lifted (kg)'], c=data['Weight Class'].astype('category').cat.codes, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.title('Amount Lifted vs Age Colored by Weight Class')
plt.colorbar(scatter, label='Weight Class')
plt.show()
```

```python
sns.histplot(data, x='Amount Lifted (kg)', hue='Lift Type', multiple='stack', palette='viridis')
plt.title('Distribution of Amount Lifted by Lift Type')
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(data=data, x='Weight Class', y='Amount Lifted (kg)', hue='Lift Type', palette='dark')
plt.title('Average Amount Lifted by Weight Class and Lift Type')
plt.xlabel('Weight Class')
plt.ylabel('Average Amount Lifted (kg)')
plt.show()
```

```python
sns.boxplot(data=data, x='Age', y='Weight Class', hue='Lift Type', palette='Set1')
plt.title('Boxplot of Age Lifted by Weight Class')
plt.xlabel('Weight Class')
plt.ylabel('Age')
plt.show()
```

```python
sns.boxplot(data=data, x='Weight Class', y='Amount Lifted (kg)')
plt.title('Boxplot of Amount Lifted by Weight Class')
plt.xlabel('Weight Class')
plt.ylabel('Amount Lifted (kg)')
plt.show()
```

```python
sns.boxplot(data=data, x='Weight Class', y='Amount Lifted (kg)', hue='Lift Type', palette='coolwarm')
plt.title('Boxplot of Amount Lifted by Weight Class and Lift Type')
plt.xlabel('Weight Class')
plt.ylabel('Amount Lifted (kg)')
plt.legend(title='Lift Type')
plt.show()
```

```python
sns.boxplot(data=data, x='Age', y='Amount Lifted (kg)', hue='Weight Class', palette='muted')
plt.title('Boxplot of Amount Lifted by Age and Weight Class')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.show()
```

```python
sns.countplot(data=data, x='Lift Type', hue='Weight Class', palette='pastel')
plt.title('Distribution of Lift Types Across Weight Classes')
plt.xlabel('Lift Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.scatterplot(data=data, x='Age', y='Amount Lifted (kg)', hue='Weight Class', palette='Set1')
plt.title('Scatter Plot of Amount Lifted vs Age')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.show()
```

```python
sns.stripplot(data=data, x='Weight Class', y='Amount Lifted (kg)', hue='Lift Type', dodge=True, palette='Paired')
plt.title('Strip Plot of Amount Lifted by Weight Class and Lift Type')
plt.xlabel('Weight Class')
plt.ylabel('Amount Lifted (kg)')
plt.show()
```

```python
corr = data[['Age', 'Amount Lifted (kg)']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', square=True)

plt.title('Heatmap of Correlation Between Age and Amount Lifted')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(data[['Age', 'Amount Lifted (kg)', 'Weight Class']], hue='Weight Class')
plt.suptitle('Pair Plot of Age, Amount Lifted, and Weight Class', y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(data, col='Lift Type', hue='Weight Class', palette='husl', col_wrap=3)
g.map(sns.scatterplot, 'Age', 'Amount Lifted (kg)')
g.add_legend()
plt.show()
```
